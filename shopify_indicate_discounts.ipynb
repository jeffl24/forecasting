```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.options.display.max_columns = 999

def try_float(x):
    try:
        return str(format(float(x), '.0f'))
    except ValueError as e:
        return x

shopify_data = pd.read_csv('orders_export_20180620.csv', 
                           low_memory=False, 
                           dtype={'Lineitem sku': 'str', 'Name': 'str'},
                           parse_dates = ['Paid at', 'Created at'])
```

```python
sku_master = pd.read_excel("C:/Users/limzi/OneDrive/Forecasting & Reporting/PCSG Master List.xlsx", 
                           sheetname = 'SKU List',
                           dtype = {'SKU': 'str'})#.dropna(subset= ['Barcode']).iloc[:, :-7]
sku_master['Barcode'] = sku_master['Barcode'].apply(try_float).astype('str')

sku_master[sku_master['SKU'] == '6110']
```

```python
shopify_data_merge_msrp = shopify_data.merge(sku_master[['SKU', 'SG Retail']], how = 'left', left_on='Lineitem sku', right_on='SKU')
shopify_data_merge_msrp.rename(columns = {'SKU': 'Ref SKU', 'SG Retail': 'MSRP'}, inplace = True)
shopify_data_merge_msrp['Discount Amount'] = shopify_data_merge_msrp.MSRP - shopify_data_merge_msrp['Lineitem price']

new_columns = ['Name', 'Email', 'Financial Status', 'Paid at', 'Fulfillment Status',
       'Fulfilled at', 'Accepts Marketing', 'Currency', 'Subtotal', 'Shipping',
       'Taxes', 'Total', 'Discount Code', 'Shipping Method',
       'Created at', 'Lineitem quantity', 'Lineitem name', 'Lineitem price',
       'Lineitem compare at price', 'Lineitem sku',
       'Lineitem requires shipping', 'Lineitem taxable',
       'Lineitem fulfillment status', 'Shipping City', 'Shipping Zip', 'Shipping Province',
       'Shipping Country', 'Shipping Phone',
       'Cancelled at', 'Payment Method', 'Payment Reference',
       'Refunded Amount', 'Vendor', 'Outstanding Balance', 'Employee',
       'Location','Ref SKU', 'MSRP', 'Discount Amount']

shopify_data_merge_msrp = shopify_data_merge_msrp[new_columns]

shopify_data_merge_msrp['discount_percentage'] = shopify_data_merge_msrp['Discount Amount'] / shopify_data_merge_msrp['MSRP']
shopify_data_merge_msrp
#shopify_data_merge_msrp.sort_values('Created at', ascending = True)
```

```python
sales_by_email_sku = pd.DataFrame(post_gc_launch_sales.groupby(['Email', 'Name', 'Lineitem sku', 'Lineitem quantity'])['Lineitem price'].sum()).reset_index()
sales_by_email_sku['total_gc_amount'] = sales_by_email_sku['Lineitem quantity'] * sales_by_email_sku['Lineitem price']

total_gc_sold = pd.DataFrame(sales_by_email_sku[sales_by_email_sku['Lineitem sku'].str.contains('gc')].groupby('Email')['total_gc_amount'].sum())
total_gc_sold.reset_index(inplace = True)
```

```python
expense_by_email = pd.DataFrame(post_gc_launch_sales.groupby(['Email', 'Name', 'Payment Method'])['Total'].sum()).reset_index()
expense_by_email = expense_by_email[expense_by_email['Payment Method'].str.contains('Gift')]
expense_gc_only = pd.DataFrame(expense_by_email.groupby('Email')['Total'].sum()).reset_index()
```

```python
total_gc_sold_deduct = total_gc_sold.merge(expense_gc_only[['Email','Total']], how = 'left', on='Email')
total_gc_sold_deduct['Remaining'] = total_gc_sold_deduct['total_gc_amount'] - total_gc_sold_deduct['Total']
total_gc_sold_deduct['Remaining'].fillna(total_gc_sold_deduct['total_gc_amount'], inplace=True)
total_gc_sold_deduct
```
