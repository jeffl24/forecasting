```python
import numpy as np
import pandas as pd
pd.options.display.max_columns = 999
```

```python
lock_in = pd.read_excel('ForecastModellingPython.xlsx', sheet_name='LockIn', dtype={'SKU': 'str'})
lock_in = lock_in.melt(id_vars=['SKU', 'Product Name'], var_name='Month', value_name='lock_qty')
lock_in.head()
```

```python
sales_ets = pd.read_excel('ForecastModellingPython.xlsx', sheet_name='SalesQtyETS', dtype={'SKU': 'str'})
sales_ets = sales_ets.melt(id_vars=['SKU', 'Product Name'], var_name='Month', value_name='sales_qty')
sales_ets.head()
```

```python
inventory = pd.read_excel('ForecastModellingPython.xlsx', sheet_name='EOM1', dtype={'SKU': 'str'})
inventory = inventory.melt(id_vars=['SKU', 'Product Name'], var_name='Month', value_name='inv_qty')
inventory.head()
```

```python
top_off = pd.read_excel('ForecastModellingPython.xlsx', sheet_name='Top Off Order', dtype={'SKU': 'str'})
top_off = top_off.melt(id_vars=['SKU', 'Product Name'], var_name='Month', value_name='topoff_qty')
top_off.head()
```

```python
lock_sales = pd.merge(lock_in, sales_ets[['SKU', 'Month', 'sales_qty']], how='left', on = ['SKU', 'Month'])
```

```python
lock_sales_eom = pd.merge(lock_sales, inventory[['SKU', 'Month', 'inv_qty']], how='left', on = ['SKU', 'Month'])
```

```python
lock_sales_eom_topoff = pd.merge(lock_sales_eom, top_off[['SKU', 'Month', 'topoff_qty']], how='left', on = ['SKU', 'Month'])
```
